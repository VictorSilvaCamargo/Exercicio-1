```python
import math
import matplotlib.pyplot as plt

def altura_telhado(comprimento):
    altura = comprimento * 0.25
    return altura


def desenhar_triangulo(altura,comprimento):
    # Calcular os três pontos do triângulo
    x1, y1 = 0, 0
    x2, y2 = comprimento, 0
    x3, y3 = comprimento,altura
    
    # Plotar o triângulo
    plt.plot([x1, x2], [y1, y2], 'k')
    plt.plot([x2, x3], [y2, y3], 'k')
    plt.plot([x3, x1], [y3, y1], 'k')
    
    # Configurar o gráfico
    plt.xlim([-1, 8])
    plt.ylim([-1, altura+1])
    plt.title('Telhado com alturaa minima de {} metros'.format(altura))
    plt.grid(True)
    plt.show()

for comprimento in range(2,8):
  altura = altura_telhado(comprimento)
  desenhar_triangulo(altura,comprimento)
```

```python
import matplotlib.pyplot as plt
import math

plt.style.use("default")

def circulo_trigonometrico(x, y):
    # cria plot
    fig, axes = plt.subplots()

    # cria círculo trigonométrico
    c = plt.Circle((0, 0), radius=1.0, edgecolor='black', facecolor='none')
    axes.add_artist(c)

    # configura eixos
    axes.axhline(y=0, color='black')
    axes.axvline(x=0, color='black')

    # define triângulo
    axes.plot([0, x], [0, 0], color='blue', label='cateto adjacente')
    axes.plot([x, x], [0, y], color='green', label='cateto oposto')
    axes.plot([0, x], [0, y], color='red', label='hipotenusa')

    # desenha triângulo retângulo
    theta = math.atan2(y, x)
    hipotenusa = math.sqrt(x**2 + y**2)
    adjacente = math.cos(theta) * hipotenusa
    oposto = math.sin(theta) * hipotenusa
    triangulo = plt.Polygon([(0,0), (adjacente, 0), (adjacente, oposto)], 
                            edgecolor='purple', facecolor='none')
    axes.add_artist(triangulo)

    # limites do meu gráfico
    if x > 0:
        plt.xlim(-1.1, 1.1)
    else:
        plt.xlim(-1.1, 0.1)
    if y > 0:
        plt.ylim(-1.1, 1.1)
    else:
        plt.ylim(-1.1, 0.1)

    # renderiza gráfico
    plt.legend()
    plt.title("Círculo Trigonométrico")
    plt.show()

print("Informe o angulo")
grau=int(input())

graus_radianos = math.radians(grau)
x = math.cos(graus_radianos)
y = math.sin(graus_radianos)


circulo_trigonometrico(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def num_int():
    input_valido = False
    while not input_valido:
        n_numero = input("Digite um número inteiro entre 1 e 5: ")
        if n_numero.isdigit():
            n = int(n_numero)
            if 1 <= n <= 5:
                input_valido = True
            else:
                print("Valor inválido. Tente novamente.")
        else:
            print("Valor inválido. Tente novamente.")
    
    # cria um vetor com os ângulos em radianos
    x = np.linspace(0, 2*np.pi*n, 1000)
    
    # calcula as funções trigonométricas
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = np.tan(x)
    
    # plota os gráficos
    fig, (li1, li2, li3) = plt.subplots(3, 1, figsize=(10, 10))
    li1.plot(x, y1, color='blue')
    li1.set_title('Seno')
    li2.plot(x, y2, color='green')
    li2.set_title('Cosseno')
    li3.plot(x, y3, color='red')
    li3.set_title('Tangente')
    plt.show()

num_int()
```
